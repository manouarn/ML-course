```python
import matplotlib.pyplot as plt
import numpy as np
import sklearn
```

```python
import pandas as pd
# here you put your directory instead /home/anouar/Documents/cours/ml/tp/data/d
df=pd.read_csv('/home/anouar/Documents/cours/ml/tp/data/dataTP2.csv')
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(df.X, df.y, s=10)
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.show()
```

```python
X=df.iloc[:,:1].values
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
plt.figure(figsize=(8, 6))
plt.scatter(X_train, y_train,s=50, c='red', label="Training Data")
plt.scatter(X_test, y_test, s=50, c='blue', label="Testing Data")
plt.legend()
plt.show()
```

```python
from sklearn import linear_model #  1.importer le modele dont le nom est name_of_mode
model = linear_model.LinearRegression() #2.Instancier le modèle
model.fit(X_train, y_train) #3.Ajuster le modèle linéaire.
print("Train Accuracy:",model.score(X_train, y_train)) #4.Afficher l’erreur de l’entrainement (Training error ) 
print("Test Accuracy:",model.score(X_test, y_test))#5.Afficher l’erreur de teste (Testing error) 
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_train, y_train,s=50, c='red', label="Training Data")
plt.scatter(X_test, y_test, s=50, c='blue', label="Testing Data")
plt.xlabel("$x$")
plt.ylabel("$y$")
x_min, x_max = plt.xlim()
nx = 100
xx = np.linspace(x_min, x_max, nx).reshape(-1,1)
plt.plot(xx,model.predict(xx), color='g', label="Droite de regression linéaire")
plt.title("Modele de regression linéaire")
plt.legend()
plt.show()
```

```python
y_pred=model.predict(X_test)
from sklearn import metrics
print("Accuracy:",metrics.mean_squared_error(y_test, y_pred))
```

```python
y_train_pred=model.predict(X_train)
print("Accuracy:",metrics.r2_score(y_train, y_train_pred))
```

```python
print("Accuracy:",model.score(X_train, y_train))
```

```python
print("Accuracy:",metrics.mean_squared_error(X_test, y_test))
```
